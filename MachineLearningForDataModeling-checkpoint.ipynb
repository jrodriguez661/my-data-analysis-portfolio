## I. BMI Analysis 

```python
x = df["bmi"].values.reshape(-1, 1)
y = df["target"].values.reshape(-1,1)
```

#### Machine Learning: Test and Train Methodology
- ``x``           = This represents the feature matrix (independent variable, predictors)
- ``y``           = This represents the target variable (dependent variable, responses)
- ``test_size``   = Specifies that <b>30% of the data<b> will be used for testing, and the remaining <b>70% for training.<b>
-``random_state`` = Ensures that the split is reproducible. Using the same ``random_state`` will always produce the same split.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.30, random_state=100)

print("Original size of x: ", x.size)
print("Size of x_test es: ", x_train.size)
print("Size of x_train es: ", x_test.size)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(x_train,y_train)
y_train_predict = reg.predict(x_train)
y_test_predict = reg.predict(x_test)

print("y_train_predict: ", y_train_predict.size)
print("y_test_predict: ", y_test_predict.size)
```

#### BMI Linear Regression Visualization


```python
min_bmi = df['bmi'].min()
max_bmi = df['bmi'].max()
print (min_bmi, max_bmi)
```

```python
#Graphing the Visualization
plt.scatter(x_train, y_train, c = "blue", alpha = 0.09027530)
plt.scatter(x_test, y_test, c = "orange", alpha = 0.170552)
x_plot = np.linspace(0, max_bmi).reshape(-1, 1)
y_plot = reg.predict(x_plot)
plt.plot(x_plot, y_plot,"r--", lw = 4)

#Labels and Titles
plt.title("Linear Regression Analysis: BMI vs. Disease Progression")
plt.xlabel("BMI (Body Mass Index)")
plt.ylabel("Target (Disease Progression)")

#Annotating points along the regression line for predictions
for i in range(0, len(x_plot), 20):
    plt.annotate(f"{y_plot[i][0]:.2f}", (x_plot[i], y_plot[i]), fontsize=10, color="black")


plt.show()
```

#### BMI $R^2$ Scores Analysis
<p>There are a couple of observations to be made regarding the $R^2$ scores: they're both relatively close, so the model <b>is not overfitting</b>. However, both scores are relatively low, meaning the model is likely <b>underfitting</b>, meaning it isn't capturing enough complexity in the relationship.</p>

```python
from sklearn.metrics import r2_score
print("Training", r2_score(y_train, y_train_predict))
print("Test", r2_score(y_test, y_test_predict))
```

## II. Average Blood Pressure (BP)

```python
x = df["bp"].values.reshape(-1, 1)
y = df["target"].values.reshape(-1,1)
```

#### Machine Learning: Test and Train Methodology
- ``x``           = This represents the feature matrix (independent variable, predictors)
- ``y``           = This represents the target variable (dependent variable, responses)
- ``test_size``   = Specifies that <b>30% of the data<b> will be used for testing, and the remaining <b>70% for training.<b>
-``random_state`` = Ensures that the split is reproducible. Using the same ``random_state`` will always produce the same split.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.30, random_state=100)

print("Original size of x: ", x.size)
print("Size of x_test es: ", x_train.size)
print("Size of x_train es: ", x_test.size)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(x_train,y_train)
y_train_predict = reg.predict(x_train)
y_test_predict = reg.predict(x_test)

print("y_train_predict: ", y_train_predict.size)
print("y_test_predict: ", y_test_predict.size)
```

```python
min_bp = df['bp'].min()
max_bp = df['bp'].max()
print (min_bp, max_bp)
```

#### BP Linear Regression Visualization

```python
#Graphing the Visualization
plt.scatter(x_train, y_train, c = "blue", alpha = 0.09027530)
plt.scatter(x_test, y_test, c = "orange", alpha = 0.170552)
x_plot = np.linspace(0, max_bp).reshape(-1, 1)
y_plot = reg.predict(x_plot)
plt.plot(x_plot, y_plot,"g--", lw = 4)

#Labels and Titles
plt.title("Linear Regression Analysis: Average Blood Pressure vs. Disease Progression")
plt.xlabel("Blood Pressure")
plt.ylabel("Target (Disease Progression)")

#Annotating points along the regression line for predictions
for i in range(0, len(x_plot), 20):
    plt.annotate(f"{y_plot[i][0]:.2f}", (x_plot[i], y_plot[i]), fontsize=10, color="black")


plt.show()
```

#### BP $R^2$ Scores Analysis
<p>Similar to the $R^2$ from the BMI analysis, they're both relatively close, therefore, <b>not overfitting</b>. However, both scores are relatively low as well, meaning the model is likely <b>underfitting</b>.</p>

```python
from sklearn.metrics import r2_score
print("Training", r2_score(y_train, y_train_predict))
print("Test", r2_score(y_test, y_test_predict))
```
